```python
import requests
import json
import time
import datetime
import pandas as pd
import numpy as np
```

```python
response = requests.get("http://api.lolesports.com/api/v2/highlanderTournaments?league=6")
source = response.text
id_data = json.loads(source)
```

```python
hash_json_id = []
hash_json_hash = []
f = open('LCKpara.csv','w')
header = "tournmant_Id,match_Id,description,groupName,generatedName,game_time,gameRealm,gameId,id\n"

for i in range(len(id_data['highlanderTournaments'])):
    description = id_data['highlanderTournaments'][i]['description']
    tournmant_Id = id_data['highlanderTournaments'][i]["id"]
    brackets_Id = list(id_data['highlanderTournaments'][i]['brackets'].keys())
    for j in range(len(brackets_Id)):
        match_Id = list(id_data['highlanderTournaments'][i]['brackets'][brackets_Id[j]]['matches'].keys())
        groupName = id_data['highlanderTournaments'][i]['brackets'][brackets_Id[j]]['groupName']
        for k in range(len(match_Id)):
            game_Id = list(id_data['highlanderTournaments'][i]['brackets'][brackets_Id[j]]['matches'][match_Id[k]]['games'].keys())
            hash_data = requests.get("http://api.lolesports.com/api/v2/highlanderMatchDetails?tournamentId="+tournmant_Id+"&matchId="+match_Id[k]).json()
            hash_json = hash_data['gameIdMappings']
            for m in range(len(hash_json)):
                hash_json_id.append(hash_json[m]['id'])
                hash_json_hash.append(hash_json[m]['gameHash'])
            for p in range(len(game_Id)):
                try:
                    time = id_data['highlanderTournaments'][i]['brackets'][brackets_Id[j]]['matches'][match_Id[k]]['games'][game_Id[p]]['standings']['timestamp']
                    game_time = str(datetime.datetime.fromtimestamp(time/1000).strftime('%Y-%m-%d %H:%M:%S'))
                    generatedName = str(id_data['highlanderTournaments'][i]['brackets'][brackets_Id[j]]['matches'][match_Id[k]]['games'][game_Id[p]]['generatedName'])
                    gameRealm = str(id_data['highlanderTournaments'][i]['brackets'][brackets_Id[j]]['matches'][match_Id[k]]['games'][game_Id[p]]['gameRealm'])
                    gameId = str(id_data['highlanderTournaments'][i]['brackets'][brackets_Id[j]]['matches'][match_Id[k]]['games'][game_Id[p]]['gameId'])
                    header += str(tournmant_Id)+","+str(match_Id[k])+","+str(description)+","+str(groupName)+","+generatedName+","+game_time+","+gameRealm+","+gameId+","+str(game_Id[p])+'\n'
                except Exception as ex:
                    print(ex)

df2 = pd.DataFrame({"id" : hash_json_id, "gameHash":hash_json_hash})
df2.to_csv('GameHash.csv')
f.write(header)
f.close()
```

```python
df1 = pd.read_csv('LCKpara.csv')
df2 = pd.read_csv('GameHash.csv')
df = pd.merge(df1,df2,left_on='id',right_on='id')
```
