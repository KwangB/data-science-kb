```python
import requests
import numpy as pd
import pandas as pd
import time
from datetime import datetime, timezone, timedelta
import datetime
```

```python
API_KEY  = 
challenger_league = requests.get("https://kr.api.riotgames.com/lol/league/v3/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key="+API_KEY).json()
master_league = requests.get("https://kr.api.riotgames.com/lol/league/v3/masterleagues/by-queue/RANKED_SOLO_5x5?api_key="+API_KEY).json()
challenger_master_name = []
for i in range(len(challenger_league['entries'])):
    challenger_master_name.append(challenger_league['entries'][i]['playerOrTeamName'])
for i in range(len(master_league['entries'])):
    challenger_master_name.append(master_league['entries'][i]['playerOrTeamName'])

    
accountId = []
summoner_body_name = []
for i in range(len(challenger_master_name)):
    summoner_body = requests.get('https://kr.api.riotgames.com/lol/summoner/v3/summoners/by-name/'+str(challenger_master_name[i])+"?api_key="+API_KEY).json()
    time.sleep(3)
    try:
        accountId.append(summoner_body['accountId'])
        summoner_body_name.append(summoner_body['name'])
        print(summoner_body['accountId'],summoner_body['name'])
    except Exception as ex:
        print(ex,i,challenger_master_name[i])
```

```python
df = pd.DataFrame({'accountId':accountId,'summoner_name': summoner_body_name})
df.to_csv('id_df.csv')
```
