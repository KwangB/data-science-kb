```python
import requests
import json
from pandas.io.json import json_normalize
```

```python
response = requests.get('https://acs.leagueoflegends.com/v1/stats/game/TRLH3/1001380191/timeline?gameHash=43a46f49938bacca')
time_line_data = json.loads(response.text)
```

```python

first_result = json_normalize(time_line_data['frames'][0]['participantFrames']['1'])
first_result['time'] = time_line_data['frames'][0]['timestamp']
for i in range(2,11):
    result2 = json_normalize(time_line_data['frames'][0]['participantFrames'][str(i)])
    first_result = first_result.append(result2,sort=True)
    
for i in range(1,len((time_line_data['frames']))):
    for j in range(1,11):
        result2 = json_normalize(time_line_data['frames'][i]['participantFrames'][str(j)])
        result2['time'] = time_line_data['frames'][i]['timestamp']
        first_result = first_result.append(result2,sort=True)
```

```python
first_event = json_normalize(time_line_data['frames'][0]['events'])
first_event['time']= time_line_data['frames'][0]['timestamp']
for i in range(1,len((time_line_data['frames']))):
    result2 = json_normalize(time_line_data['frames'][i]['events'])
    result2['time'] = time_line_data['frames'][i]['timestamp']
    first_event = first_event.append(result2,sort=True)
```

```python
first_event.to_csv('event.csv')
```
